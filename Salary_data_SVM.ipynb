```python
data_ = raw_data.copy()
data_.drop('Salary',axis=1, inplace =True)
data_ = pd.get_dummies(data_.iloc[:,:])
data_.head()
```

```python
data_['Salary'] = raw_data.Salary
data_.head()
```

label encoding
---

```python
#Label encoding salary
le = LabelEncoder()
le.fit(data_["Salary"])
data_["Salary"]=le.transform(data_["Salary"])
data_.head()
```

```python
# Label Encoding of categrical variables
df= raw_data.copy()

#loading encoder
label_encoder = LabelEncoder()

df["education"]=label_encoder.fit_transform(df["education"])
df["workclass"]=label_encoder.fit_transform(df["workclass"])
df["maritalstatus"]=label_encoder.fit_transform(df["maritalstatus"])
df["sex"]=label_encoder.fit_transform(df["sex"])
df["race"]=label_encoder.fit_transform(df["race"])
df["occupation"]=label_encoder.fit_transform(df["occupation"])
df["relationship"]=label_encoder.fit_transform(df["relationship"])
df["native"]=label_encoder.fit_transform(df["native"])
df['Salary'] = np.where(df['Salary'].str.contains(" >50K"), 1, 0)


df.head(10)
```

Feature Selections
---

```python
x = df.drop('Salary',axis=1)
y = df['Salary']
```

Spliting Dataset : TEST/TRAIN
---

-  StratifiedKFold

```python
from collections import Counter
# Create StratifiedKFold object.
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=1)

# enumerate the splits and summarize the distributions
for train_ix, test_ix in skf.split(x, y):
# select rows
    train_X, test_X = x.iloc[train_ix], x.loc[test_ix]
    train_y, test_y = y.iloc[train_ix], y.iloc[test_ix]
# summarize train and test composition
counter_train = Counter(train_y)
counter_test = Counter(test_y)
print('Training Data',counter_train,'Testing Data',counter_test)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1,stratify=y)
```

```python
# summarize train and test composition
counter_train = Counter(y_train)
counter_test = Counter(y_test)
print('Training Data',counter_train,'Testing Data',counter_test)
```

```python
print("Shape of X_train: ",x_train.shape)
print("Shape of X_test: ", x_test.shape)
print("Shape of y_train: ",y_train.shape)
print("Shape of y_test",y_test.shape)
```

 #  Support Vector Classifier Using Linear Kernel

SVC(kernel='linear', probability=True)

```python
# initiating the model
model = SVC(kernel = "linear",random_state=40,gamma=0.1,C=1.0)
# fit the model
model.fit(x_train, y_train)
```

# model analysis


```python
# Predict the results
y_pred = model.predict(x_test)
y_pred
```

```python
# Check accuracy score

from sklearn.metrics import accuracy_score

print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

Here, y_test are the true class labels and y_pred are the predicted class labels in the test-set.

```python
# Compare the train-set and test-set accuracy
y_pred_train = model.predict(x_train)

y_pred_train
```

```python
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train)))
```

Check for overfitting and underfitting

```python
# print the scores on training and test set

print('Training set score: {:.4f}'.format(model.score(x_train, y_train)))
print('Test set score: {:.4f}'.format(model.score(x_test, y_test)))
```

The training-set accuracy score is 0.81 while the test-set accuracy to be 0.80.
These two values are quite comparable. So, there is no sign of overfitting

```python
# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
# Print the Confusion Matrix and slice it into four pieces
print('Confusion matrix\n\n', cm)
print('\nTrue Positives(TP) = ', cm[0,0])
print('\nTrue Negatives(TN) = ', cm[1,1])
print('\nFalse Positives(FP) = ', cm[0,1])
print('\nFalse Negatives(FN) = ', cm[1,0])
```

```python
# visualize confusion matrix with seaborn heatmap

cm_matrix = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'],
                                 index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='YlGnBu')
```

```python
# Classification metrices
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
# Classification accuracy

TP = cm[0,0]
TN = cm[1,1]
FP = cm[0,1]
FN = cm[1,0]
```

```python
# classification accuracy

classification_accuracy = (TP + TN) / float(TP + TN + FP + FN)
print('Classification accuracy : {0:0.4f}'.format(classification_accuracy))
```

```python
# Classification error

classification_error = (FP + FN) / float(TP + TN + FP + FN)
print('Classification error : {0:0.4f}'.format(classification_error))
```

```python
# Precision

precision = TP / float(TP + FP)
print('Precision : {0:0.4f}'.format(precision))
```

```python
# Recall

recall = TP / float(TP + FN)
print('Recall or Sensitivity : {0:0.4f}'.format(recall))
```

```python
#true positive rate

true_positive_rate = TP / float(TP + FN)
print('True Positive Rate : {0:0.4f}'.format(true_positive_rate))
```

```python
# False Positive Rate

false_positive_rate = FP / float(FP + TN)
print('False Positive Rate : {0:0.4f}'.format(false_positive_rate))
```

```python
# Specificity

specificity = TN / (TN + FP)
print('Specificity : {0:0.4f}'.format(specificity))
```
